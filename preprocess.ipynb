### **2.** Features’ transformation

Some features are not being represented in the most usable format, hence certain transformations are necessary, for instance:
* As stated above, BirthYear, NKids and NTeens are stored as floating points, whereas it makes more sense to be integers. This step will be performed after the treatment of missing values, since it is not possible to convert empty spaces, i.e. NaN values or infinite values to integer.

##### **2.1** Converting CustomerFrom to CustomerDays

```python
# Overrides CustomerFrom after pd.to_datetime(), changing its data type from object to datetime
data["CustomerFrom"] = pd.to_datetime(data["CustomerFrom"], format="%Y/%m/%d")  
print("'CustomerFrom' data type:", data["CustomerFrom"].dtype)
```

It is not enough to simply convert CustomerFrom into datetime format. In order to include it in our analysis, the feature needs to be converted into a directly measurable quantity. That is, the number of days between the date stored in the feature and some reference date.

```python
# Setup reference date as the first day of 2018
reference_date = "2018/01/01"
# Computing the number of days since the reference date 
customer_days = (pd.to_datetime(reference_date, format="%Y/%m/%d") - data["CustomerFrom"]).dt.days 
# Inserting the new column (CustomerDays)
data.insert(loc=4, column="CustomerDays", value=customer_days)

# Deleting the 'BirthYear' feature, as it is being replaced by Age, that contained the same information
del data["CustomerFrom"]

# Checking the data
data.head()
```

#### **3.2** Missing values

When a dataset presents missing values one can either remove those records or impute. Traditionally, the missing values in the numerical columns are imputed with central tendency measures.

As previously observed, the columns that have missing values are:
* BirthYear, replaced by Age;
* Education;
* NKids;
* NTeens;
* Income.

It is understandable that some of these features can be associated with sensitive topics and therefore some people choose not to share this information. For example, people with the highest and the lowest incomes may decide not to disclose it.

<br>

**3.2.1** To have a better overview of the missing values it is created a dataframe in which the missing values are represented as True and the non-missing values as False.

```python
# Considers empty strings and numpy.inf as missing values
pd.options.mode.use_inf_as_na = True

# It is created a dataframe (data_na) in which the missing values are represented as True and the remaning as False 
data_na = data.isna()

# Visualize data_na DataFrame
display("DataFrame with the missing values as True:", data_na.tail())
```

<br>

Using the previous dataframe, we can count the amount of missing values that exist in each column (considering only the columns that have at least one). The boolean values True and False are also defined as the number 1 and 0, respectively. Therefore, the sum of the values in each row (0 if it is False and 1 if True), of a specific column, results in the amount of missing values in that column.

To count those missing values is used the function missing_values(), which performs the respective sum and stores the result in a dataframe in an absolute and relative format.

```python
def missing_values(df, use_inf_as_na=True):       
    pd.options.mode.use_inf_as_na = use_inf_as_na
    # Counts the number of missing values in each row and stores it in na_count
    data_na = data.isna().sum() 
    # Identifies columns with at least one missing values
    features_na = data_na[data_na > 0]    
    # Returns a dataframe with the columns with missing values and corresponding amount and proportion
    data_na = pd.DataFrame.from_dict({"Number of missing values": features_na, 
                                    "Proportion of missing values": features_na.divide(df.shape[0])})
    return data_na

# Using the missing_values 
missing_values(data, True)
```

By using this function it is clear that the Income (which has 51 missing values) is the feature with the largest amount of missing values and BirthYear with the least (having 22 missing values) among the features that contain this values.

<br>

**3.2.2** To overcome the problem of missing values, it is used the ``.fillna()`` method. In the code below:
1.  For each numeric column, imputes the missing values with the mean of the respective column;
2.  For each categorical column, imputes the missing values with the mode of the respective column;

```python
# Using the the missing_values function on the dataframe data
display("Before imputing:", missing_values(data))

# Replacing the missing values by the mean for numerical columns
data["Age"].fillna(value=data["Age"].mean(), inplace=True)
data["NKids"].fillna(value=data["NKids"].mean(), inplace=True)
data["NTeens"].fillna(value=data["NTeens"].mean(), inplace=True)
data["Income"].fillna(value=data["Income"].mean(), inplace=True)

# Replacing the missing values by the mode for the categorical column
data["Education"].fillna(value=data["Education"].mode().values[0], inplace=True)

# Calling missing_values function to check if there is still missing values
display("After imputing:", missing_values(data))
```

The tables are meant to show the amount of missing values and its proportion to the number of rows of the column, as previously displayed. However, these two tables show those same values before and after imputation. Thus, after imputation, we can observe that all the empty spaces were filled, not remaining any missing values.

##### **3.3.2** Eliminating extreme value from income

Considering that the minimum value of Income is 1730 and the maximum is 99999999, it is relevant to check if this last value is actually an isolated case. To accomplish that, we will perform some conditions on the Income feature.

```python
display("Income higher than 200 000:",data[data["Income"]>200000])
```

In the table above we have the rows of the dataframe 'data' in which the Income feature is larger than 200 000. There are two rows, corresponding to the income values 666666.0 and 99999999.0. Those values are both large considering the quartiles of this feature. However, 99999999.0 is significantly larger than 666666.0, which makes it a considerable extreme of this column. Hence, it is benefitial for the model that this value is removed.

To be able to remove rows easily from now on, it is created a function (remove_rows) that makes the process simpler. This function has two parameters, the column and the value. The column corresponds to the feature, from which we intend to remove a specific value. The parameter value is that specific value that is intended to be removed.

```python
def remove_rows(Column, value):
    # Getting the row of the value to be removed 
    index = data.loc[(data[Column] == value), :].index
    # Removing row
    data.drop(index, inplace=True)
```

```python
# Using the function remove_rows() to remove the row in which Income has the extreme value
remove_rows('Income', 99999999.0)
```

To check if the code is working we can build the previous table again.

```python
display("Income higher than 200 000:",data[data["Income"]>200000])
```

### **4.** Identification and treatment of inconsistent facts

Over the process of studying and transforming the features of the 'data' dataset, it was possible to recognize some facts that need to be satisfied among some features. For instance: 

* For an individual to have a value different from 0 in Revenue, it would have to respond to the campaign. Hence, if the feature Response is 0, the Revenue would have to be 0 as well, and if it the Response is 1, the Revenue is 11 (which corresponds to the monetary units received by each customer that positively responded to the campaign). Given the fact that every single value of Revenue is 11, the solution would be to replace the value 11 by 0, when Response was equal to 0. However, it was observed that the feature Revenue had a null standard deviation and therefore, was not useful for the model. If that solution was actually put in practise, Revenue would provide exactly the same information as Response, which would not be helpful. Thus, no change is needed in regards to the independent variable mentioned.
* The MntGoldProds variable should not be larger than the amount of money spent among all the products (that is, the sum of the variables MntFishProducts, MntMeatProducts, MntFruits, MntSweetProducts, MntWines).
* If an individual has never purchased a product from the company (NWebPurchases, NCatalogPurchases and NStorePurchases are equal to 0), then it does not make sense for Recency to have a numerical value. Instead, it should be filled with a 'Not Applicable' value. 
* If NWebPurchases, NCatalogPurchases and NStorePurchases are equal to 0, it is conflicting to have values larger than 0 on the amount spent in products (MntMeatProducts, MntWines, MntFruits, MntFishProducts, MntSweetProducts)). Therefore, in those cases, the values of the last five features mentioned, that are larger than 0, should be replaced by 0.
* The NDealsPurchases should be null when no purchases are made, that is, when NWebPurchases, NCatalogPurchases and NStorePurchases are equal to 0. This is because the purchases represented by NDealsPurchases should be made in one of the three possible ways, that is, through the store, the website or a catalog.

<br>
